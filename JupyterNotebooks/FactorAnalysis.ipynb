# Factor Analysis

```python
%matplotlib widget
import pandas as pd
import numpy as np
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
import common
from Data import Merge
```

```python
data = Merge().data

# remove unused columns
data = data.drop(['LDELTOTAL', 'DIGITSCOR'], axis=1) 

dataLabels = data.columns.values
dataNumpy = data.to_numpy()
rowsRemove = []
### Remove data incomplete
for i in range(len(dataNumpy)):
    count = 0
    for j in dataNumpy[i]:
        if j != '':
            count += 1
    #if count < len(dataLabels) - 2:
    if count < 32:
        rowsRemove.append(i)

data = data.drop(rowsRemove)
```

```python
fa = FactorAnalyzer(rotation="varimax").fit(data)
```

```python
loadings = fa.loadings_
eigenvector, eigenvalues = fa.get_eigenvalues()
```

## Scree Plot

```python
plt.close()

xvals = range(1, data.shape[1]+1)
plt.scatter(xvals, eigenvector)
plt.plot(xvals, eigenvector)
plt.title("Scree Plot")
plt.xlabel("Factor")
plt.ylabel("Eigenvalue")
plt.grid()
plt.show()
```
