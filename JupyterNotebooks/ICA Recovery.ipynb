Recover ICAs from csv file

```python
def recover(data: pd.DataFrame) -> pd.DataFrame:
    # Get names for rows and columns
    rows = []
    cols = []
    
    for i, j in [x.split(' -> ') for x in list(data.Labels.unique())]:
        rows.append(i)
        cols.append(j)
    
    # Remove fields repeated
    rows = list(dict.fromkeys(rows))
    cols = list(dict.fromkeys(cols))
    
    # Generate Data Frame
    df = pd.DataFrame(index=rows, columns=cols)
    
    # Populating Data Frame
    for _, row in data.iterrows():
        place = row[1].split(' -> ')
        df.at[place[0], place[1]]=row[2]
        
    # Filing empty and returning
    return df.fillna(0).astype("float")
```

```python
filtred = recover(pd.read_csv("Data/local/Correlações filtradas.csv"))
unFiltred = recover(pd.read_csv("Data/local/Correlações sem filtro.csv"))
```

Function to apply filtred ICs on a given row

```python
def applyFiltered(row: pd.DataFrame):
    data_filtred = pd.DataFrame(columns=filtred.columns.values, index=filtred.T.columns.values)
    for i in filtred.columns.values:
        for j in filtred.T.columns.values:
            data_filtred[i][j] = filtred[i][j] * row[i]
    return data_filtred      
```
