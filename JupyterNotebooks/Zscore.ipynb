```python
import pandas as pd
from scipy.stats import zscore
import common
```

```python
corr = common.loadFile("ICADataCorrelation")
data = common.loadFile("CleanedData").drop(["RID", "VISCODE", "AGE", "PTEDUCAT", "PTGENDER"], axis=1)
```

```python
score = pd.DataFrame()
for i in data.columns.values:
    score[i] = zscore(data[i].astype('int'))
score = score.drop("DX", axis=1)
```

```python
score
```

```python
import math
dataY = data.loc[:,['DX']].to_numpy().astype('int').flatten()
avg=sum(dataY)/len(dataY)

def calcSig(data):
    mean = sum(data)/len(data)
    dev =[(x - mean) ** 2 for x in data]
    sig = math.sqrt(sum(dev)/len(data))
    return sig

sig = calcSig(dataY)
print("Representation for DX Z Score:")
for i in range(3):
    print(f"Z score for {i} is {(i-avg)/sig}")
```

```python
def corrScore(score, corr):
    return f"{corr} 'X' {score}"
```

```python

result = pd.DataFrame(index=corr.T.columns.values, columns=score.T.columns.values)
for i in score.T:
    for j in corr.T:
        result[i][j] = corrScore(i, j)
```
